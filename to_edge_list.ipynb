```python
import csv

# Load the data from a CSV file
with open('readyDatset.csv', 'r') as f:
    reader = csv.reader(f)
    data = list(reader)
# Extract the source and destination nodes for each edge
edges = [(row[4], row[3]) for row in data]
```

```python
import networkx as nx
import pandas as pd
def create_graph_from_dataframe(dataframe: pd.DataFrame):
  # first create a network x representation of the graph from the pandas dataframe
  match_network_x_graph = nx.from_pandas_edgelist(
      dataframe, source="source", target="insiderthreat"
  )
  return match_network_x_graph
  
```

```python
df = pd.read_csv("readyDatset.csv")
graph=create_graph_from_dataframe(dataframe=df)
```

```python
import matplotlib.pyplot as plt
nx.draw(graph, with_labels=True, )
plt.show()
```
